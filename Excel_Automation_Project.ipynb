```python
# Step 1: Upload Excel file
from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]

# Step 2: Read Excel
import pandas as pd

df = pd.read_excel(file_name)

print("Original Data:")
display(df.head())

# Step 3: Data Cleaning
df_cleaned = df.drop_duplicates()
df_cleaned = df_cleaned.fillna(0)

# Step 4: Column-wise Summary
numeric_cols = df_cleaned.select_dtypes(include=['number'])

summary = pd.DataFrame({
    'Total': numeric_cols.sum(),
    'Average': numeric_cols.mean(),
    'Count': numeric_cols.count(),
    'Missing Values': df[numeric_cols.columns].isnull().sum()
})

print("\nSummary:")
display(summary)

# Step 5: Create Chart
import matplotlib.pyplot as plt

plt.figure()
numeric_cols.sum().plot(kind='bar')
plt.title("Total by Numeric Columns")
plt.xlabel("Columns")
plt.ylabel("Total")
plt.xticks(rotation=45)
plt.tight_layout()

chart_file = "chart.png"
plt.savefig(chart_file)
plt.close()

# Step 6: Save formatted Excel
output_file = "final_output.xlsx"


with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    df_cleaned.to_excel(writer, sheet_name='Cleaned Data', index=False)
    summary.to_excel(writer, sheet_name='Summary')

    workbook  = writer.book
    worksheet1 = writer.sheets['Cleaned Data']
    worksheet2 = writer.sheets['Summary']

    # Format headers
    header_format = workbook.add_format({
        'bold': True,
        'border': 1
    })

    # Apply header format
    for col_num, col_name in enumerate(df_cleaned.columns):
        worksheet1.write(0, col_num, col_name, header_format)

    for col_num, col_name in enumerate(summary.columns):
        worksheet2.write(0, col_num + 1, col_name, header_format)

    # Auto column width
    for i, col in enumerate(df_cleaned.columns):
        max_len = max(df_cleaned[col].astype(str).map(len).max(), len(col))
        worksheet1.set_column(i, i, max_len + 2)

    for i, col in enumerate(summary.columns):
        worksheet2.set_column(i+1, i+1, 18)

    # Insert chart image
    worksheet2.insert_image('H2', chart_file)

# Step 7: Download final file
files.download(output_file)
```
